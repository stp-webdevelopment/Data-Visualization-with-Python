# Part 1: Data Visualizations and Analysis
## TASK 1.1: Line Chart of Automobile Sales Over Time
Using pandas to plot a line chart showing how automobile sales fluctuate from year to year.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (you can replace this with your dataset)
data = {'Year': [2000, 2001, 2002, 2003, 2004],
        'Sales': [150000, 160000, 170000, 165000, 180000]}

df = pd.DataFrame(data)

# Plotting the line chart
plt.figure(figsize=(8, 5))
plt.plot(df['Year'], df['Sales'], marker='o', color='b')
plt.title('Automobile Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

## TASK 1.2: Line Chart for Different Vehicle Types
### Plotting different lines for categories of vehicle types to observe sales trends.

```python
# Sample data with vehicle types (replace with actual data)
data = {'Year': [2000, 2001, 2002, 2003, 2004],
        'Sedan': [50000, 52000, 53000, 51000, 55000],
        'SUV': [40000, 42000, 43000, 44000, 46000],
        'Truck': [30000, 31000, 32000, 33000, 34000]}

df = pd.DataFrame(data)

# Plotting the line chart for different vehicle types
plt.figure(figsize=(8, 5))
plt.plot(df['Year'], df['Sedan'], marker='o', label='Sedan', color='r')
plt.plot(df['Year'], df['SUV'], marker='o', label='SUV', color='g')
plt.plot(df['Year'], df['Truck'], marker='o', label='Truck', color='b')
plt.title('Sales Trends by Vehicle Type')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

## TASK 1.3: Visualization with Seaborn to Compare Sales During Recession and Non-Recession Periods
### Using Seaborn to compare vehicle type sales trends during recession vs non-recession periods.

```python
import seaborn as sns

# Sample data indicating recession and non-recession periods (replace with actual data)
data = {'Year': [2000, 2001, 2002, 2003, 2004],
        'VehicleType': ['Sedan', 'SUV', 'Truck', 'Sedan', 'SUV'],
        'Sales': [50000, 40000, 30000, 55000, 46000],
        'Recession': [False, True, True, False, False]}

df = pd.DataFrame(data)

# Using Seaborn to compare trends during recession and non-recession periods
plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x='Year', y='Sales', hue='Recession', style='VehicleType')
plt.title('Sales Trend During Recession vs Non-Recession Periods')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

## TASK 1.4: Subplot to Compare GDP Trends During Recession and Non-Recession Periods
### Develop line plots to compare GDP trends during recession vs non-recession periods.

```python
# Sample GDP data for recession and non-recession periods
data = {'Year': [2000, 2001, 2002, 2003, 2004],
        'GDP_Recession': [12000, 11500, 11300, 11800, 12100],
        'GDP_NonRecession': [12200, 12400, 12600, 12800, 13000]}

df = pd.DataFrame(data)

# Subplot for GDP comparison during recession and non-recession periods
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].plot(df['Year'], df['GDP_Recession'], marker='o', color='r')
axes[0].set_title('GDP During Recession Period')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('GDP')

axes[1].plot(df['Year'], df['GDP_NonRecession'], marker='o', color='g')
axes[1].set_title('GDP During Non-Recession Period')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('GDP')

plt.tight_layout()
plt.show()
```

## TASK 1.5: Bubble Plot to Display the Impact of Seasonality on Automobile Sales
### A bubble plot to show seasonality impact on sales.

```python
# Sample data for seasonality impact
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
        'Sales': [20000, 22000, 24000, 25000, 23000],
        'SeasonalityFactor': [0.8, 0.85, 0.9, 0.95, 0.8]}

df = pd.DataFrame(data)

# Bubble plot
plt.figure(figsize=(8, 5))
plt.scatter(df['Month'], df['Sales'], s=df['SeasonalityFactor']*1000, c='red', alpha=0.5)
plt.title('Impact of Seasonality on Automobile Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

# TASK 2.1: Create a Dash Application with a Meaningful Title 
### Start by creating the basic Dash application structure. Set a meaningful title for the application.

```python
# TASK 2.1: Create a Dash Application with a Meaningful Title 
### Start by creating the basic Dash application structure. Set a meaningful title for the application.
import dash
from dash import dcc, html
import plotly.express as px
import pandas as pd

# Sample data for demonstration purposes
df = pd.DataFrame({
    'Year': [2000, 2001, 2002, 2003, 2004],
    'Sales': [150000, 160000, 170000, 165000, 180000],
    'Recession': [False, True, True, False, False]
})

# Initialize the Dash app
app = dash.Dash(__name__)

# Set a meaningful title
app.title = "Automobile Sales and Economic Trends Dashboard"

# Layout of the app
app.layout = html.Div([
    html.H1("Automobile Sales and Economic Trends Dashboard")
])

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

# TASK 2.2: Add Drop-downs to Your Dashboard with Appropriate Titles and Options 
### You can use dcc.Dropdown to add dropdowns for selecting different options. For example, selecting vehicle types or recession periods.

```python
# Add dropdowns to the layout
app.layout = html.Div([
    html.H1("Automobile Sales and Economic Trends Dashboard"),
    
    # Dropdown for vehicle types
    dcc.Dropdown(
        id='vehicle-type-dropdown',
        options=[
            {'label': 'Sedan', 'value': 'Sedan'},
            {'label': 'SUV', 'value': 'SUV'},
            {'label': 'Truck', 'value': 'Truck'}
        ],
        value='Sedan',  # default value
        style={'width': '50%'}
    ),
    
    # Dropdown for recession period
    dcc.Dropdown(
        id='recession-period-dropdown',
        options=[
            {'label': 'Recession', 'value': 'True'},
            {'label': 'Non-Recession', 'value': 'False'}
        ],
        value='True',  # default value
        style={'width': '50%'}
    )
])
```

## TASK 2.3: Add a Division for Output Display with Appropriate 'id' and 'classname' Property 

```python
app.layout = html.Div([
    html.H1("Automobile Sales and Economic Trends Dashboard"),
    
    dcc.Dropdown(
        id='vehicle-type-dropdown',
        options=[
            {'label': 'Sedan', 'value': 'Sedan'},
            {'label': 'SUV', 'value': 'SUV'},
            {'label': 'Truck', 'value': 'Truck'}
        ],
        value='Sedan',
        style={'width': '50%'}
    ),
    
    dcc.Dropdown(
        id='recession-period-dropdown',
        options=[
            {'label': 'Recession', 'value': 'True'},
            {'label': 'Non-Recession', 'value': 'False'}
        ],
        value='True',
        style={'width': '50%'}
    ),
    
    # Division to display output
    html.Div(id='output-container', className='output-container', style={'margin-top': '20px'})
])
```

## TASK 2.4: Creating Callbacks; Define the Callback Function to Update the Input Container

```python
from dash import Input, Output

@app.callback(
    Output('output-container', 'children'),
    [Input('vehicle-type-dropdown', 'value'),
     Input('recession-period-dropdown', 'value')]
)
def update_output(vehicle_type, recession_period):
    filtered_df = df[df['Recession'] == (recession_period == 'True')]  # Filter based on recession period
    total_sales = filtered_df[filtered_df['Year'] == 2004]['Sales'].values[0]  # Example condition
    
    return f"Selected Vehicle Type: {vehicle_type}, Recession Period: {recession_period}, Total Sales in 2004: {total_sales}"
```

## TASK 2.5: Create and Display Graphs for Recession Report Statistics

```python
@app.callback(
    Output('recession-graph', 'figure'),
    [Input('vehicle-type-dropdown', 'value'),
     Input('recession-period-dropdown', 'value')]
)
def update_recession_graph(vehicle_type, recession_period):
    filtered_df = df[df['Recession'] == (recession_period == 'True')]
    
    fig = px.line(filtered_df, x='Year', y='Sales', title='Recession Report: Sales Trend')
    return fig
```

## TASK 2.6: Create and Display Graphs for Yearly Report Statistics

```python
@app.callback(
    Output('yearly-graph', 'figure'),
    [Input('vehicle-type-dropdown', 'value')]
)
def update_yearly_graph(vehicle_type):
    filtered_df = df[df['Sales'] > 150000]  # Example condition for vehicle type filtering
    
    fig = px.line(filtered_df, x='Year', y='Sales', title='Yearly Report: Sales Trend')
    return fig
```
